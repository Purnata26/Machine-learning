```
import os,cv2
import numpy as np
data_path='/content/drive/My Drive/Face Mask Dataset/Train'
catagory=os.listdir(data_path)
labels=np.arange(len(catagory))
label_dict=dict(zip(catagory,labels))
label_dict
```

```
test_data_path='/content/drive/My Drive/Face Mask Dataset/Validation'
test_catagory=os.listdir(test_data_path)
test_label=np.arange(len(test_catagory))
label_dict_test=dict(zip(test_catagory,test_label))
label_dict_test
```

```
img_size=100
target=[]
all_img=[]

for i in catagory:
  file_path=os.path.join(data_path,i)
  img_names=os.listdir(file_path)
  for j in img_names:
    img_path=os.path.join(file_path,j)
    img=cv2.imread(img_path)
    try:

      gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
      #covert img in gray scale
      resized=cv2.resize(gray,(img_size,img_size))
      #resize image to 100,100
      all_img.append(resized)
      target.append(label_dict[i])
    
      
    except Exception as e:
      print('Exception:',e)

all_img                
```

```
img_size=100
test_images=[]
test_image_label=[]
for i in test_catagory:
  folder=os.path.join(test_data_path,i)
  types=os.listdir(folder)
  for j in types:
    image_path=os.path.join(folder,j)
    image=cv2.imread(image_path)
    try:
      gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
      resized=cv2.resize(gray,(img_size,img_size))
      test_images.append(resized)
      test_image_label.append(label_dict_test[i])
    except Exception as e:
      print('exeption')
test_images
    

```

```

test_images=np.array(test_images)/255
test_images=np.reshape(test_images,(test_images.shape[0],img_size,img_size,1))
test_image_label=np.array(test_image_label)
from keras.utils import np_utils
test_image_label=np_utils.to_categorical(test_image_label)
print(test_images.shape)
print(test_image_label.shape)
```

```
all_img=np.array(all_img)/255
all_img=np.reshape(all_img,(all_img.shape[0],img_size,img_size,1))
target=np.array(target)
from keras.utils import np_utils
new=np_utils.to_categorical(target)
all_img.shape
```

```
new.shape
```

```
data=all_img
data.shape
target=new
```

```
from keras.models import Sequential
from keras.layers import Activation,Dense,Flatten,Dropout,Conv2D,MaxPooling2D
from keras.callbacks import ModelCheckpoint
model=Sequential()
model.add(Conv2D(32,(3,3),input_shape=data.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dropout(0.2))
model.add(Dense(64,activation='relu'))
#Dense layer of 64 neurons
model.add(Dense(2,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```
model.summary()
```

```
model.fit(data,target,epochs=3,batch_size=1,validation_split=0.1)
```

```
model.predict(test_images,batch_size=1)
```

```
model.evaluate(test_images,test_image_label,batch_size=1)
```
