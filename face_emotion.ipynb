```
import pandas as pd
data=pd.read_csv('/content/drive/My Drive/Colab Notebooks/train.csv')
data.head()
```

```
pix=data.pixels.str.split(expand=True)
pix




```

```
data=data.drop('pixels',axis=1)
data
```

```
train_data=pd.concat([data,pix],axis=1)
train_data
```

```
data1=pd.read_csv('/content/drive/My Drive/Colab Notebooks/test.csv')
data1
```

```
test_data=data1.pixels.str.split(expand=True)
test_data
```

```
train_data=train_data.values
```

```
trainX = train_data[:, 1:].reshape(train_data.shape[0],48,48,1).astype( 'float32' )

trainX
```

```
trainX.shape
```

```

X_train = trainX / 255.0
X_train
```

```
from keras.utils import np_utils
y_train = train_data[:,0]
y_train=np_utils.to_categorical(y_train)
y_train
```

```
y_train.shape
```

```
test_data=test_data.values
```

```
test_X=test_data[:,0:].reshape(test_data.shape[0],48,48,1).astype('float32')
test_X
```

```
test_X.shape
```

```
from keras.models import Sequential
from keras.layers import Activation,Dense,Flatten,MaxPooling2D,Dropout,Conv2D
from keras.callbacks import ModelCheckpoint
model=Sequential()
model.add(Conv2D(200,(3,3),input_shape=X_train.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(100,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(100,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dropout(0.2))
model.add(Dense(64,activation='relu'))
#Dense layer of 64 neurons
model.add(Dense(7,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```
checkpoint = ModelCheckpoint('model-{epoch:03d}.model',monitor='val_loss',verbose=0,save_best_only=True,mode='auto')

model.fit(X_train,y_train,epochs=5,batch_size=1,callbacks=[checkpoint],validation_split=0.2)
```
